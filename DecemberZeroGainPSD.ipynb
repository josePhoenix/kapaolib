By default, the "overall" slope computed when the slope data are read in is a median value across all subapertures, whether they are well-illuminated or not (see `Slopes.__init__`). If we have the intensity map for the run (and we should!), then we can do better.

The `IntensityMap` class computes a set of "good" subapertures as follows: it takes the median value subaperture-wise for each of the 97 subapertures, then applies a constraint based on either the `TOP_N_SUBAPERTURES` value (default: 50) or the `min_light` keyword argument to the IntensityMap constructor (default: None). In the `TOP_N_SUBAPERTURES` case, the "good" subapertures are merely the top 50 by median value. If `min_light` is supplied, any subaperture with a median value across the run that is greater than `min_light` is included in the "good" subapertures.

The instance attribute `IntensityMap.include_mask` is a boolean array with length 97 (the number of subapertures) that is true where the subaperture is 'good' to include in computations and false otherwise. In the code below, the mask is applied to the slope X and Y data before taking the median, which results in an improved overall slope time series (`slope_x.new_overall` and `slope_y.new_overall`).

```python
slope_x.new_overall = np.median(slope_x.data[:,intensity_map.include_mask], axis=1)
slope_y.new_overall = np.median(slope_y.data[:,intensity_map.include_mask], axis=1)

figure(figsize=(10,5))
plot(slope_x.overall_slopes, label="Old Overall")
plot(slope_x.new_overall, label="Masked Overall")
xlabel("Time step")
ylabel("X slope")
legend()

figure(figsize=(10,5))
plot(slope_y.overall_slopes, label="Old Overall")
plot(slope_y.new_overall, label="Masked Overall")
xlabel("Time step")
ylabel("Y slope")
legend()
```

## Compute the FFT with the better-overall-subtracted data

The `Slope` class computes a power spectrum after reading in the slopes, but if you've done additional processing to the data you can pass in arguments to its `do_fft` method and compute it yourself:

    def do_fft(self, data=None, subtracted_data=None, fromidx=0, toidx=None)

This is what we do in the next two cells for a `subtracted_x_data` array and a `subtracted_y_data` array that contain the slope data with the overall trend removed *not* by overall medianning but *instead* by the masked median process above.

```python
subtracted_x_data = slope_x.data - slope_x.new_overall[:,np.newaxis]
x_freqs, x_power, x_subtracted_power = slope_x.do_fft(subtracted_data=subtracted_x_data)
```

```python
subtracted_y_data = slope_y.data - slope_y.new_overall[:,np.newaxis]
y_freqs, y_power, y_subtracted_power = slope_y.do_fft(subtracted_data=subtracted_y_data)
```

```python
figure(figsize=(6,5))

subplot(211)
plot(x_freqs, np.average(slope_x.power[:,intensity_map.include_mask], axis=1), label="Average Power, Good Subaps (X)")
plot(x_freqs, np.average(x_subtracted_power[:,intensity_map.include_mask], axis=1), label="Residual Power, Good Subaps (X)")
plot(x_freqs, x_freqs**(-2.0/3.0) * 10.0, label="$f\,^{-2/3}$")
plot(x_freqs, x_freqs**(4.0/3.0) * 100.0, label="$f\,^{4/3}$")
xlim(np.min(x_freqs[x_freqs > 0]), np.max(x_freqs))
ylim(1, 500)
xscale('log', nonposy='mask')
yscale('log', nonposy='mask')
ylabel('Power (slope data units)')
legend(fontsize=10)

subplot(212)
plot(y_freqs, np.average(slope_y.power[:,intensity_map.include_mask], axis=1), label="Average Power, Good Subaps (Y)")
plot(y_freqs, np.average(y_subtracted_power[:,intensity_map.include_mask], axis=1), label="Residual Power, Good Subaps (Y)")
plot(y_freqs, y_freqs**(-2.0/3.0) * 10.0, label="$f\,^{-2/3}$")
plot(y_freqs, y_freqs**(4.0/3.0) * 100.0, label="$f\,^{4/3}$")
xlim(np.min(y_freqs[y_freqs > 0]), np.max(y_freqs))
ylim(1, 500)
xscale('log', nonposy='mask')
yscale('log', nonposy='mask')
xlabel('Frequency (Hz)')
ylabel('Power (slope data units)')
legend(fontsize=10)

tight_layout()
```
